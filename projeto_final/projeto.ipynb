```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import unicodedata
import numpy as np
```

```python
df=pd.read_csv("pokemon.csv")
df.info()
```

### Sera que o tamanho dos pokémons afetam diretamente em seu peso ?
Na vida real isso não é nem uma pergunta,obiviamente a altura afeta o seu peso,mas sera que no universo de pokémon isto tambem ocorre?

```python
df["height"] = df["height"].astype(str).str.replace(r'm', '', regex=True).str.strip()
df["weight"] = df["weight"].astype(str).str.replace(r'kg', '', regex=True).str.strip()

df["height"]=pd.to_numeric(df["height"])
df["weight"]=pd.to_numeric(df["weight"])
df=df.dropna(subset=["height","weight"])
```

```python
plt.bar(df["height"],df["weight"],color="red")
plt.title("A relação entre o peso e a altura de um pokémon")
plt.xlabel("Altura")
plt.ylabel("peso")
plt.show()
```

Isso é uma causalidade
,pois existe sim uma relação entre o peso e a altura de um pokémon,isto fica perceptivel ao observar o grafico acima.
Entretanto existem varias essesões,um dos maiores exemplos disto dentro do universo pokémon é o pokémon Cosmoem que tem 10cm e mesmo assim pesando 999,9kg.
Então é perceptivel que boa parte dos pokémons tem um peso dentro da logica,mas alguns pokémons seguem uma logica ficticia.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import unicodedata
import numpy as np
```

```python
df=pd.read_csv("dataset_lazaro.csv")
df.info()


df["horas_estudo"]=pd.to_numeric(df["horas_estudo"])
df["nota"]=pd.to_numeric(df["nota"])
df=df.dropna(subset=["horas_estudo","nota"])

def remover_acentos(texto):
    if isinstance(texto,str):
        normalized=unicodedata.normalize("NFKC",texto)
        return normalized.encode("ascii","ignore").decode("utf-8")
    return texto
#Retirando acentose cedilhas dos dados
for coluna in df.columns:
    df[coluna]=df[coluna].apply(remover_acentos)
```

```python
horas_estudo=[1,2,3,4,5,6,2,4,5,7,8,3,6,7,8]
notas=[4,5,6,7,8,10,3,6,7,8,9,8,9,5,9]
plt.scatter(horas_estudo,notas)
#calcula os coeficientes da linha de tendencia
coeficientes=np.polyfit(horas_estudo,notas,1)
funcao_tendencia=np.poly1d(coeficientes)
#Plota a linha de tendencia
plt.plot(horas_estudo,funcao_tendencia(horas_estudo),
        color="blue",linestyle='--' )

plt.title("horas de estudo vs nota")
plt.xlabel("Horas de estudo")
plt.ylabel("Nota")
plt.show()
```

```python
#Grafico de exemplo 2
categorias=["A","B","C","D","E","F"]
precos=[10,25,15,32,12,43]
custos=[8,19,9,26,5,100]
plt.bar(categorias,precos,color="blue")
plt.bar(categorias,custos,color="green")
plt.title("relacao entre custo e preco ")
plt.xlabel("categoria")
plt.ylabel("valor")
plt.show()
```
